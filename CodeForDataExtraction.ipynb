```python
import json
import pandas as pd

with open('Main#1.json', 'r', encoding='utf-8') as json_file: 
    data = json.load(json_file)
    json_file.close()
    


# reading the csv
# csv_file = open('ExtractedData#2.csv', 'w', encoding='utf-8', newline='')
# csv_file.close()
data1 = pd.read_csv('ExtractedData#2.csv', encoding='utf-8') 
# data1.head()
counter = 4498
tweet_data = data['tweets'] 
for tweet in tweet_data: 
    tweet_id = tweet['id']
    description = tweet['user']['description']
    followers_count = tweet['user']['followers_count']
    friends_count = tweet['user']['friends_count']
    listed_count = tweet['user']['listed_count']
    favourites_count = tweet['user']['favourites_count']
    statues_count = tweet['user']['statuses_count']
   
    print(counter)
    full_text = ''
    expanded_URL = ''
    url = ''
    
    if 'retweeted_status' in tweet:
#         print("*",len (tweet['retweeted_status']['entities']['urls']))
        if tweet['retweeted_status']['truncated'] == False:
            full_text = tweet['text']
            if len (tweet['retweeted_status']['entities']['urls']) == 0:
                 None
            else:    
                url = tweet['retweeted_status']['entities']['urls'][0]['url']
                expanded_URL = tweet['retweeted_status']['entities']['urls'][0]['expanded_url'] 
        
        else:
            full_text = tweet['retweeted_status']['extended_tweet']['full_text']
            url = tweet['retweeted_status']['entities']['urls'][0]['url']
            expanded_URL = tweet['retweeted_status']['entities']['urls'][0]['expanded_url'] 
            
    else:
        full_text = tweet['text']
        if len(tweet['entities']['urls'])==0:
            None
        else:
            expanded_URL =  tweet['entities']['urls'][0]['expanded_url']
            url = url = tweet['entities']['urls'][0]['url']
            full_text = tweet['text']
            
     
           
    
   
    
#     #-----------------------------------------------------------------------------------------------------
    
   
    data1.loc [counter, 'id'] = tweet_id
    data1.loc [counter, 'text'] = full_text
    data1.loc [counter, 'url'] = url
    data1.loc [counter, 'description'] = description
    data1.loc [counter, 'followers_count'] = followers_count
    data1.loc [counter, 'friends_count'] = friends_count
    data1.loc [counter, 'listed_count'] = listed_count
    data1.loc [counter, 'favourites_count'] = favourites_count
    data1.loc [counter, 'statues_count'] = statues_count
    data1.loc [counter, 'expanded_URL'] = expanded_URL
    
    data1.to_csv('ExtractedData#2.csv',index=False, encoding='utf-8-sig')
    counter = counter + 1
#     if counter >=4499:
#         break
   
    
    
 
```
